```python
#Making our pre_BPE train data
with open("/content/gdrive/MyDrive/ArabicGEC/QALB-data/QALB-2015-L2-Train.cor", "r") as correct_answers1, open("/content/gdrive/MyDrive/ArabicGEC/QALB-data/QALB-2014-L1-Train.cor", "r") as correct_answers2, open("data/train/train.preBPE.cor", "w") as train_correct, open("/content/gdrive/MyDrive/ArabicGEC/QALB-data/QALB-2015-L2-Train.sent", "r") as original_sents1, open("/content/gdrive/MyDrive/ArabicGEC/QALB-data/QALB-2014-L1-Train.sent", "r") as original_sents2, open("data/train/train.preBPE.incor", "w") as train_incorrect:
  for line1, line2 in zip(correct_answers1, original_sents1):
    good_line1 = line1.split()[1:]
    text1 = " ".join(good_line1).strip()
    text1 = re.sub(r'(\s\W)(?=\1)', '', text1)

    good_line2 = line2.split()[1:]
    text2 = " ".join(good_line2).strip()
    text2 = re.sub(r'(\s\W)(?=\1)', '', text2)
    if text1.count(".") < 5 and text2.count(".") <5:
      print(text1, file=train_correct)
      print(text2, file=train_incorrect)


  for line1, line2 in zip(correct_answers2, original_sents2):
    good_line1 = line1.split()[1:]
    text1 = " ".join(good_line1).strip()
    text1 = re.sub(r'(\s\W)(?=\1)', '', text1)

    good_line2 = line2.split()[1:]
    text2 = " ".join(good_line2).strip()
    text2 = re.sub(r'(\s\W)(?=\1)', '', text2)
    if text1.count(".") < 5 and text2.count(".") <5:
      print(text1, file=train_correct)
      print(text2, file=train_incorrect)
```

```python
#the preprocessing step
! fairseq-preprocess --source-lang incor --target-lang cor \
  --tokenizer moses \
  --trainpref data/train/train.tok \
  --validpref data/dev/dev.tok \
  --testpref  data/test/test.tok \
  --destdir data-bin/arabic.tokenized.incor-cor \
  --thresholdsrc 2 \
  --thresholdtgt 2 
```

```python
#!pip install tensorboardX
```

```python
#the training step with Alla's reccomendations!!!!
#origianlly used --- transformer_vaswani_wmt_en_de_big --- before trying mbar_large

! fairseq-train data-bin/arabic.tokenized.incor-cor \
--arch transformer_vaswani_wmt_en_de_big \
--max-epoch 110 \
--max-tokens 5000 \
--optimizer adam \
--adam-betas '(0.9,0.98)' \
--dropout 0.3 \
--clip-norm 1.0 \
--criterion label_smoothed_cross_entropy \
--label-smoothing 0.1 \
--log-format simple \
--log-interval 100 \
--warmup-updates 8000 \
--lr-scheduler inverse_sqrt \
--warmup-init-lr '1e-07' \
--lr 0.0005 \
--stop-min-lr '1e-09' \
--update-freq 16 \
--keep-last-epochs 10 \
--ddp-backend=no_c10d \
--patience 3 \
--amp
```
